<span style="font-size:25px;">**Data Visualisation using Various Graphs**</span>


```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Selling_Price"], y=df["Present_Price"], alpha=0.7)
plt.xlabel("Selling Price")
plt.ylabel("Present Price")
plt.title("Actual vs Present Car Prices")
plt.show()
```

```python
data1 = df.iloc[0:100]
sns.barplot(y='Selling_Price', x='Car_Name', data = data1)
plt.xticks(rotation= 90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x=df["Year"], y=df["Selling_Price"], marker="o", color="b")
plt.title("Selling Price Trends Over the Years", fontsize=14, fontweight="bold")
plt.xlabel("Year", fontsize=12, fontweight="bold")
plt.ylabel("Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Present_Price"], y=df["Selling_Price"], alpha=0.7)
plt.title("Actual vs. Predicted Car Prices", fontsize=14, fontweight="bold")
plt.xlabel("Present Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.ylabel("Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["Fuel_Type"], y=df["Selling_Price"], palette="coolwarm")
plt.title("Impact of Fuel Type on Selling Price", fontsize=14, fontweight="bold")
plt.xlabel("Fuel Type", fontsize=12, fontweight="bold")
plt.ylabel("Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=df["Transmission"], y=df["Selling_Price"], palette="viridis")
plt.title("Transmission Type vs. Selling Price", fontsize=14, fontweight="bold")
plt.xlabel("Transmission Type", fontsize=12, fontweight="bold")
plt.ylabel("Average Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Driven_kms"], y=df["Selling_Price"], alpha=0.5, color="g")
plt.title("Kilometers Driven vs. Selling Price", fontsize=14, fontweight="bold")
plt.xlabel("Kilometers Driven", fontsize=12, fontweight="bold")
plt.ylabel("Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["Owner"], y=df["Selling_Price"], palette="Set2")
plt.title("Ownership Impact on Selling Price", fontsize=14, fontweight="bold")
plt.xlabel("Number of Owners", fontsize=12, fontweight="bold")
plt.ylabel("Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=df["Selling_type"], y=df["Selling_Price"], palette="magma")
plt.title("Selling Type vs. Selling Price", fontsize=14, fontweight="bold")
plt.xlabel("Selling Type", fontsize=12, fontweight="bold")
plt.ylabel("Average Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["Year"], y=df["Selling_Price"], palette="Set1")
plt.title("Yearly Trends in Selling Price", fontsize=14, fontweight="bold")
plt.xlabel("Year", fontsize=12, fontweight="bold")
plt.ylabel("Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x=df["Transmission"], y=df["Selling_Price"], palette="Accent")
plt.title("Distribution of Selling Price by Transmission Type", fontsize=14, fontweight="bold")
plt.xlabel("Transmission Type", fontsize=12, fontweight="bold")
plt.ylabel("Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["Present_Price"], bins=20, kde=True, color="red")
plt.title("Distribution of Present Prices", fontsize=14, fontweight="bold")
plt.xlabel("Present Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.ylabel("Count", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["Selling_Price"], bins=20, kde=True, color="blue")
plt.title("Distribution of Selling Prices", fontsize=14, fontweight="bold")
plt.xlabel("Selling Price (in Lakhs)", fontsize=12, fontweight="bold")
plt.ylabel("Count", fontsize=12, fontweight="bold")
plt.grid(True)
plt.show()
```
