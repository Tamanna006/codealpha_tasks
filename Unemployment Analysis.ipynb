<span style="font-size:25px"><b>Data Visualisation using Various Graphs</b></span>



```python
import seaborn as sns
import matplotlib.pyplot as plt
df.columns = df.columns.str.strip()
data1 = df.iloc[0:500]
sns.barplot(x='Region', y='Estimated Unemployment Rate (%)', data=data1)
plt.xticks(rotation=90) 
plt.show()
```

```python
data1 = df.iloc[0:500]  
plt.figure(figsize=(15, 6))  
sns.boxplot(x='Region', y='Estimated Employed', data=data1)
plt.xticks(rotation=90)  
plt.show()
```

```python
data1 = df.iloc[0:500]  
plt.figure(figsize=(15, 6))  
sns.violinplot(x='Region', y='Estimated Employed', data=data1)
plt.xticks(rotation=90)  
plt.show()
```

<span style="font-size:25px"><b>Which City Has Highest Labour Participation Rate??</b></span>


```python
sns.lineplot(x='Region', y='Estimated Labour Participation Rate (%)', data=data1)
plt.xticks(rotation=90)  
plt.show()
```

<span style="font-size:25px"><b>Which Area Has Highest Unemployment Rate??</b></span>


```python
sns.barplot(x='Area', y='Estimated Unemployment Rate (%)', data = data1)
plt.show()
```
